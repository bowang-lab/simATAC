```python
# This function calculated the ami and ari clustering metrics from the predicted labels
# given as input file.
# Input:
# file: A text file with true and predicted labels
# Output: -
def get_metrics(file):

    pred = list()
    label = list()
    
    with open(file, 'r') as fp:
        for line in fp:
            if "ground_truth" in line:
                continue
            line = line.rstrip().replace("\"", "")
            part = line.split()
            label.append(part[0])
            pred.append(part[1])
            
    nmi = normalized_mutual_info_score(label, pred, average_method='max')
    ami = adjusted_mutual_info_score(label, pred, average_method='arithmetic')
    ari = adjusted_rand_score(label, pred)
    
    
    print('nmi: ',str(nmi))
    print('ami: ',str(ami))
    print('ari: ',str(ari))

```
